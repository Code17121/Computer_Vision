```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((300,300,3),dtype = np.uint8)
plt.imshow(img)
```

```python
img[100:200,100:200] = [0,255,0]
plt.imshow(img)
```

# Resize the image

```python
img_resized = cv2.resize(img,(15,150),interpolation = cv2.INTER_LINEAR)
plt.imshow(img_resized)
```

# Cropping

```python
cropped_image = img[50:250,50:250]
plt.imshow(cropped_image)
```

# Roatation

```python
center = (img.shape[0]//2,img.shape[0]//2)
matrix = cv2.getRotationMatrix2D(center,45,1.0)
img_roatated = cv2.warpAffine(img,matrix,(img.shape[1],img.shape[0]))

plt.imshow(img_roatated)
```

# Flipping

```python
img_flipped = cv2.flip(img,1) # horizontal flip
plt.imshow(img_flipped)
```

# Translation

```python
matrix = np.float32([ [1, 0, 50], [0, 1, 50]])
img_translated = cv2.warpAffine(img,matrix,(img.shape[1],img.shape[0]))

plt.imshow(img_translated)
```
