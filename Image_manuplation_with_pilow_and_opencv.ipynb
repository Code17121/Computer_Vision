```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python

path = '/content/anime.webp'
```

```python
image = Image.open(path)
```

```python
image.format
```

```python
image.size
```

```python
image.mode
```

```python
plt.imshow(image)
plt.title('anime_girl')
plt.axis('off')
plt.show()
```

```python
resize_image = image.resize((200,200))
plt.imshow(resize_image)
plt.title('anime_girl')
plt.axis('off')
plt.show()
```

```python
cropped_image = image.crop((100,200,200,300))
plt.imshow(cropped_image)
plt.title('anime_girl')
plt.axis('off')
plt.show()
```

```python
gray_image = image.convert('L')
plt.imshow(gray_image)
plt.title('anime_girl')
plt.axis('off')
plt.show()
```

```python
rotate_image = image.rotate(45)
plt.imshow(rotate_image)
plt.title('anime_girl')
plt.axis('off')
plt.show()
```

```python
#Everythin is just in one cell

fig,axes = plt.subplots(1,4,figsize = (16,4))
axes[0].imshow(resize_image)
axes[0].set_title('resized_image')
axes[1].imshow(cropped_image)
axes[1].set_title('cropped_image')
axes[2].imshow(rotate_image)
axes[2].set_title('rotated_image')
axes[3].imshow(gray_image)
axes[3].set_title('gray_image')
plt.show()
```
