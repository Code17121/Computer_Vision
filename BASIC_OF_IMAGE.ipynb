# Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Data

```python
img = np.zeros((300,300,3),dtype = np.uint8)
plt.imshow(img)
```

```python
img1 = img.copy()
img1[100:200,100:200] = [255,0,0]
plt.imshow(img1)
```

# Reading the image


```python
cv2.imwrite('synthetic_image.png',img1)
img_read = cv2.imread('synthetic_image.png')

plt.imshow(img_read)
```

# Color Spaces

```python
gray_img = cv2.cvtColor(img1,cv2.COLOR_RGB2GRAY)
hsv_img = cv2.cvtColor(img1,cv2.COLOR_RGB2HSV)
```

```python
plt.imshow(gray_img,cmap = 'gray')
```

```python

plt.imshow(cv2.cvtColor(hsv_img, cv2.COLOR_HSV2RGB))
```

# Pixel Manuplation

```python
img1[50:100,50:100] = [0,0,255]
plt.imshow(img1)
```
