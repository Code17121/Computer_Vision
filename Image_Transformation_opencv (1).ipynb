```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.zeros((300,300,3),dtype = np.uint8)
plt.imshow(img)
```

```python
img[100:200,100:200] = [255,255,255]
plt.imshow(img)
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_img,cmap = 'gray')
```

# Blurring

```python
gauss_img = cv2.GaussianBlur(gray_img,(5,5),0)
plt.imshow(gauss_img,cmap = 'gray')
```

```python
img_median = cv2.medianBlur(gray_img,5)
plt.imshow(img_median,cmap = 'gray')
```

```python
img_bilateral = cv2.bilateralFilter(gray_img,9,75,75)
plt.imshow(img_bilateral,cmap = 'gray')
```

# Edge Detection

```python
img_sobel = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize=3)
img_sobel = cv2.convertScaleAbs(img_sobel)
img_canny = cv2.Canny(gray_img, 100, 200)
```

```python
plt.imshow(img_sobel)
```

```python
plt.imshow(img_canny)
```
